```python
# Membuat indeks untuk baris yang ingin dihapus
rows_to_drop = details_df[details_df["Amount"] <= details_df["Profit"]].index

# Menghapus baris-baris tersebut
details_df.drop(index=rows_to_drop, inplace=True)
```

```python
details_df[details_df["Amount > Profit"] == 0]
```

```python
details_df.groupby(by="Amount > Profit")["Order ID"].count()
```

```python
details_df["Margin"] = details_df["Amount"] - details_df["Profit"]
```

```python
details_df.sample(5)
```

```python
details_df.groupby(by="Category").agg({
    "Order ID": "nunique",
    "Margin": ["max", "min", "mean", "std",]
})
```

```python
details_df.groupby(by="Category").agg({
    "Order ID": "nunique",
    "Profit":  ["sum","max", "min", "mean", "std",]
})
```

```python
details_df.groupby(by="Category").agg({
    "Order ID": "nunique",
    "Quantity": ["sum","max", "min", "mean", "std",]
})
```

```python
details_df.groupby(by="PaymentMode").agg({
    "Order ID": "nunique",
})
```

```python
details_df["status"] = details_df["Order ID"].isin(order_df["Order ID"]).map({True: "Active", False: "Non Active"})
```

```python
details_df.sample(5)
```

```python
details_df.groupby(by="status")["Order ID"].count()
```

## Menggabungkan Tabel

```python
all_df = pd.merge(
    left=order_df,
    right=details_df,
    how="left",
    left_on="Order ID",
    right_on="Order ID"
)
all_df.head(10)
```

```python
all_df.groupby(by=["State", "Category"]).agg({
    "Quantity": "sum",
    "Profit": "sum"
})
```

```python
all_df.groupby(by=["Category", "Sub-Category"]).agg({
    "Quantity": "sum",
    "Profit": "sum"
})
```

```python
rata_rata = all_df.groupby('Order ID').size().mean()
print("Banyak Order rata-rata yang dilakukan per orang = ", rata_rata)
```

```python
max = all_df.groupby('Order ID').size().max()
print("Banyak Order paling banyak yang dilakukan per orang = ", max)
```

```python
min = all_df.groupby('Order ID').size().min()
print("Banyak Order paling sedikit yang dilakukan per orang = ", min)
```
